```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
# 1. Đọc dữ liệu từ file CSV
import os
# Đường dẫn khi chạy trên máy thật:
data_path = r'd:\HDFS\JOB_MARKET_BIGDATA\data\processed\Data_ITJOB_Cleaned.csv'
# Lưu ý: Đường dẫn này có thể cần đổi khi chạy trên Colab (VD: '/content/Data_ITJOB_Cleaned.csv')

if not os.path.exists(data_path):
    print('File không tồn tại. Vui lòng upload dữ liệu lên Colab và cập nhật lại data_path.')
else:
    df = pd.read_csv(data_path)
    print("Kích thước dữ liệu:", df.shape)
    display(df.head(3))
```

```python
# 2. Tiền xử lý: Kết hợp title_clean + skills_clean + description_clean
# Xử lý giá trị NaN (nếu có) bằng chuỗi rỗng
df['title_clean'] = df['title_clean'].fillna('')
df['skills_clean'] = df['skills_clean'].fillna('')
df['description_clean'] = df['description_clean'].fillna('')

df['combined_text'] = df['title_clean'] + " " + df['skills_clean'] + " " + df['description_clean']
print("Ví dụ văn bản kết hợp:", df['combined_text'].iloc[0])
```

```python
# 3. Khởi tạo mô hình và sinh Embedding
model_name = 'paraphrase-multilingual-MiniLM-L12-v2'
model = SentenceTransformer(model_name)

# Sinh embedding cho toàn bộ jobs
job_embeddings = model.encode(df['combined_text'].tolist(), show_progress_bar=True)
print("Kích thước job_embeddings:", job_embeddings.shape)
```

```python
# 4. Lưu dữ liệu và embeddings
df.to_csv('job_data.csv', index=False)
np.save('job_embeddings.npy', job_embeddings)
print("Đã lưu job_data.csv và job_embeddings.npy thành công!")
```
